# Linear Regression Analysis
Detailed academic workflow with explanations.

## Step 1: Load and Understand Data
We begin by loading the dataset and inspecting its structure.

```

import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

## Step 2: EDA
We explore distributions and relationships between variables.

```

import matplotlib.pyplot as plt
df['charges'].hist()
plt.show()
```

## Step 3: Feature Engineering
Categorical variables are encoded, and features prepared.

```

df = pd.get_dummies(df, drop_first=True)
```

## Step 4: Model Training
We train a linear regression model.

```

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df.drop('charges', axis=1)
y = df['charges']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = LinearRegression()
model.fit(X_train,y_train)
```

## Step 5: Evaluation
We evaluate performance using key metrics.

```

from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
print(r2_score(y_test,y_pred))
```
